```python
import pandas as pd
from pycaret.classification import *

# data = pd.read_csv('/home/hk_lenovo_pc/data/dataset.csv')
odata = pd.read_csv('/home/hk_sc/data/dataset.csv')
sma_pc_0 = list(odata['sma_pc_0'])
sma_pc_sum = list(odata['sma_pc_sum'])
signal = list(odata['signal'])

data = pd.DataFrame()
data['sma_pc_0'] = sma_pc_0
data['sma_pc_sum'] = sma_pc_sum
data['signal'] = signal
# shuffle the DataFrame rows
data = data.sample(frac = 1)
data.head()
```

```python
# import ClassificationExperiment and init the class
from pycaret.classification import ClassificationExperiment
exp = ClassificationExperiment()
s = setup(data, target = 'signal', session_id = 123, use_gpu=True)
exp.setup(data, target = 'signal', session_id = 123, use_gpu=True)
# compare baseline models
best = compare_models()
```

```python
# plot confusion matrix
plot_model(best, plot = 'confusion_matrix')
```

```python
# plot AUC
plot_model(best, plot = 'auc')
```

```python
# plot feature importance
plot_model(best, plot = 'feature')
```

```python
evaluate_model(best)
```
